# 3.  Codificación de variables

Conjunto de datos: titanic (incluido en seaborn)

Este conjunto de datos es excelente para trabajar con variables categóricas.

```python
# Importar librerías necesarias
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# Cargar el conjunto de datos
data = sns.load_dataset('titanic')
```

```python
# Visualizar las primeras filas
print("Conjunto de datos Titanic:")
data.head()
```

```python
# Seleccionar columnas categóricas
categorical_cols = ['sex', 'embarked', 'class']
```

```python
# Codificar con LabelEncoder
label_encoder = LabelEncoder()
data['sex_label'] = label_encoder.fit_transform(data['sex'])
```

```python
# Codificar con OneHotEncoder
onehot_encoder = pd.get_dummies(data['embarked'], prefix='embarked')
```

```python
# Combinar con el conjunto original
data = pd.concat([data, onehot_encoder], axis=1)
```

### 1. ¿Qué diferencias encuentras entre LabelEncoder y OneHotEncoder?

```python
# Comparamos las columnas transformadas
print("Resultado de LabelEncoder (sex_label):")
print(data[['sex', 'sex_label']].head())

print("\nResultado de OneHotEncoder (embarked):")
# Mostramos las columnas creadas por get_dummies
embarked_cols = [col for col in data.columns if 'embarked_' in col]
print(data[embarked_cols].head())
```

LabelEncoder asigna un número único a cada categoría (ej. male=1, female=0), lo cual es eficiente pero puede confundir al modelo haciéndole creer que hay un orden jerárquico. OneHotEncoder crea una columna nueva para cada categoría con valores binarios (0 o 1), evitando jerarquías falsas pero aumentando el número de columnas (dimensionalidad)

### 2. Crea una gráfica de barras comparando las frecuencias de 'sex' antes y después de la codificación con LabelEncoder.

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Antes de la codificación (Texto)
sns.countplot(x='sex', data=data, ax=ax1, palette='pastel')
ax1.set_title('Original: Variable "sex"')

# Después de la codificación (Números)
sns.countplot(x='sex_label', data=data, ax=ax2, palette='magma')
ax2.set_title('LabelEncoded: Variable "sex_label"')

plt.show()
```

### 3. Utiliza OneHotEncoder para codificar la columna 'class'. ¿Qué ventajas tiene este enfoque frente a LabelEncoder?

```python
# Aplicar OneHotEncoding a 'class'
class_encoded = pd.get_dummies(data['class'], prefix='class')
data = pd.concat([data, class_encoded], axis=1)

print("Nuevas columnas para 'class':")
print(class_encoded.head())
```

La ventaja es que evita que el modelo asuma que la clase 'Third' (3) es mayor o 'mejor' que la 'First' (1) por el valor numérico. Al usar columnas separadas, el algoritmo trata a cada clase como una entidad independiente sin un orden matemático implícito
