# Datos Faltantes

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.impute import SimpleImputer, KNNImputer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el conjunto de datos
diabetes = load_diabetes(as_frame=True)
data = diabetes.data
data['target'] = diabetes.target
```

```python
data.sample(10)
```

```python
  # Simular valores faltantes
np.random.seed(42)
data.loc[data.sample(frac=0.1).index, 'bmi'] = np.nan
data.loc[data.sample(frac=0.1).index, 'bp'] = np.nan
```

# Ejercicios

### 1. ¿Cuántos valores faltantes hay en cada columna?**

```python
Para esto utilizaremos los métodos isnull() y sum(). Este paso es vital para dimensionar el problema antes de aplicar cualquier técnica de limpieza.
# Contar valores nulos por columna
valores_faltantes = data.isnull().sum()
print("Valores faltantes por columna:")
print(valores_faltantes)
```

### 2. Utiliza imputación simple (media) para llenar los valores faltantes de la columna 'bmi'.

```python
La imputación simple rellena los huecos con un valor estadístico constante (como la media o mediana) de la misma columna.

from sklearn.impute import SimpleImputer

# Creamos una copia para no alterar el original y poder comparar luego
data_simple = data.copy()

# Configuramos el imputador para usar la media
imputer_media = SimpleImputer(strategy='mean')

# Aplicamos la imputación solo a la columna 'bmi'
# Nota: Reshape es necesario porque el imputer espera una matriz 2D
data_simple['bmi'] = imputer_media.fit_transform(data_simple[['bmi']])

print(f"Valores nulos en 'bmi' tras imputación simple: {data_simple['bmi'].isnull().sum()}")
```

### 3. Utiliza KNNImputer para imputar valores en las columnas 'bmi' y 'bp'. Compara los resultados con los de la imputación simple.

```python
A diferencia de la media, el KNNImputer es más inteligente: busca las filas más parecidas (vecinos) a la que tiene el dato faltante y estima el valor basándose en ellas.

from sklearn.impute import KNNImputer

# Creamos otra copia para la comparación
data_knn = data.copy()

# Configuramos el imputador (n_neighbors=5 es el estándar)
imputer_knn = KNNImputer(n_neighbors=5)

# Imputamos ambas columnas al mismo tiempo
data_knn[['bmi', 'bp']] = imputer_knn.fit_transform(data_knn[['bmi', 'bp']])

print(f"Valores nulos en 'bmi' tras KNN: {data_knn['bmi'].isnull().sum()}")
print(f"Valores nulos en 'bp' tras KNN: {data_knn['bp'].isnull().sum()}")
```

### 4. Genera un histograma comparando los datos antes y después de la imputación en la columna 'bmi'.

```python
Graficar los datos es la mejor forma de ver cómo la imputación afectó la distribución original.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

# Histograma de datos originales (con nulos omitidos por seaborn)
sns.histplot(data['bmi'], kde=True, color="blue", label="Original (con nulos)", alpha=0.5)

# Histograma tras imputación simple
sns.histplot(data_simple['bmi'], kde=True, color="red", label="Imputación Media", alpha=0.3)

# Histograma tras KNN
sns.histplot(data_knn['bmi'], kde=True, color="green", label="Imputación KNN", alpha=0.3)

plt.title("Comparativa de Distribución: bmi")
plt.legend()
plt.show()
```
